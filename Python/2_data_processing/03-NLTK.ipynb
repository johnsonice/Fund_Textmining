# <a name='tokenize'>Tokenize sentences and words
Let's use the sentence and word tokenizers and part of speech tagging on our WEO document.

```python
from nltk import sent_tokenize, word_tokenize, pos_tag, help as nltk_help
```

```python
filename = 'weo-april-2017'
```

```python
with open(filename+'.txt',encoding='utf-8') as infile:
    rawtext = infile.read() #read whole thing as ONE block
```

```python
rawtext[0:200]
```

```python
cleantext = rawtext.replace('\n',' ').replace('  ',' ').replace('- ','')
```

```python
cleantext[0:80]
```

```python
sentences = sent_tokenize(cleantext)
```

```python
sent_tokenize('Something (really?) else.')
```

```python
len(sentences)
```

```python
sentences[0]
sentences[1]
```

```python
words = word_tokenize(cleantext)
```

```python
len(words)
```

```python
words[0:10]
```

Finally, let's transform the text into an NLTK text which does a few things for us automatically.

```python
from nltk import Text as nltk_text
```

```python
finaltext = nltk_text(words)
```

```python
finaltext.vocab()
```

Now let's tag each word with the part of speech

```python
pos = pos_tag(finaltext)
```

```python
pos[0:10]
```

```python
nltk_help.upenn_tagset('NNP')
```

We've broken it up into words, so does the order matter?

```python
example_words = word_tokenize("They refuse to permit us to obtain the refuse permit")
example_pos = pos_tag(example_words)
print(example_pos)
```

```python
example_words.reverse()
print(pos_tag(example_words))
```

```python
pos_tag(word_tokenize('Buffalo buffalo buffalo Buffalo buffalo?')) # Buffalo are so confusing
```

Stemming and Lemmatization
Both are similar in that they find the root of words, but lemmatization considers context where stemming does not.

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
porter_stemmer.stem('Coffee') # change the word here to see what you get
```

Lemmatization is based on WordNet's morphy function

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmatizer.lemmatize('are','v')
```
